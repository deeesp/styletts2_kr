```python
import os
import pandas as pd
from tqdm.auto import tqdm
```

## VCTK pairing

```python
import IPython.display as ipd

vctk_root = '/Storage/speech/tts/VCTK-Corpus'
wav_vctk = os.path.join(vctk_root, 'wav48')
txt_vctk = os.path.join(vctk_root, 'txt')
train_item_list = list()
valid_item_list = list()

for spk in tqdm(os.listdir(wav_vctk)):
    l = len(os.listdir(os.path.join(wav_vctk, spk)))
    # print(l)
    l_th = 5#l*0.9
    for i, utt in enumerate(os.listdir(os.path.join(wav_vctk, spk))):
        if utt.endswith('wav'):
            utt_txt = utt[:-3]+'txt'
            txt_path = os.path.join(txt_vctk, spk, utt_txt)
            if os.path.isfile(txt_path):
                f = open(os.path.join(txt_vctk, spk, utt_txt), "r")
                text = f.read()[:-1]
                if text.endswith('"') or text.startswith('"') or text.startswith('(') or text.endswith(')'):
                    # print(text)
                    text = text.replace('"','').replace(')','',).replace('()','',)
                    print(text)
                item = {
                    "audio_path": os.path.join(wav_vctk, spk, utt),
                    "spk": spk,
                    "text": text,
                }
                if i>=l_th:
                    break
                else:
                    train_item_list.append(item)
                    
                # if i>l_th:
                #     valid_item_list.append(item)
                # else:
                #     train_item_list.append(item)
```

```python
vctk_train = pd.DataFrame.from_dict(train_item_list, orient='columns')
vctk_train.to_csv('/Storage/speech/tts/VCTK-Corpus/vctk_train_5-sample.csv', sep='|', index=False, encoding='utf-8-sig')
vctk_train
```

```python
df_valid = pd.DataFrame.from_dict(valid_item_list, orient='columns')
df_valid.to_csv('/Storage/speech/tts/VCTK-Corpus/vctk_valid.csv', sep='|', index=False, encoding='utf-8-sig')
df_valid
```
