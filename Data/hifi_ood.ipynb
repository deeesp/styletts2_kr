```python
import IPython.display as ipd

hifi_root = '/Storage/speech/tts/hi_fi_tts_v0/'
ood_test = list()

for file in tqdm(os.listdir(hifi_root)):
    if file.endswith('test.json'):
        df = pd.read_json(os.path.join(hifi_root, file), lines=True)
        ood_test.extend(list(df['text']))
```

```python
import librosa
import IPython.display as ipd
import soundfile as sf
import numpy as np
import os

hifi_root = '/Storage/speech/tts/hi_fi_tts_v0/'
hifi_item_list = []

for file in tqdm(os.listdir(hifi_root)):
    if file.endswith('train.json'):
        spk = 'hifi'+file.split('_')[0]
        print(spk)
        df = pd.read_json(os.path.join(hifi_root, file), lines=True)
        df = df[(6.0<df['duration']) & (df['duration']<10.0)][:5]
        for sample in df.iterrows():
            item = {
                "audio_path": os.path.join(hifi_root, sample[1]['audio_filepath']),
                "spk": spk,
                "text": sample[1]['text'],
            }
            # print(item['text'])
            # ipd.display(ipd.Audio(item['audio_path'], rate=24000))
            hifi_item_list.append(item)

len(hifi_item_list)
```

```python
import librosa
import IPython.display as ipd
import soundfile as sf
import numpy as np
import os

hifi_root = '/Storage/speech/tts/hi_fi_tts_v0/'
hifi_item_list = []

for file in tqdm(os.listdir(hifi_root)):
    if file.endswith('dev.json'):
        spk = 'hifi'+file.split('_')[0]
        print(spk)
        df = pd.read_json(os.path.join(hifi_root, file), lines=True)
        for sample in df.iterrows():
            item = {
                "audio_path": os.path.join(hifi_root, sample[1]['audio_filepath']),
                "spk": spk,
                "text": sample[1]['text'],
            }
            # print(item['text'])
            # ipd.display(ipd.Audio(item['audio_path'], rate=24000))
            hifi_item_list.append(item)

len(hifi_item_list)
```

```python
hifi_train = pd.DataFrame.from_dict(hifi_item_list, orient='columns')
hifi_train.to_csv('/Storage/speech/tts/hi_fi_tts_v0/hifi_train_5-sample.csv', sep='|', index=False, encoding='utf-8-sig')
hifi_train
```
